```python
import requests
from bs4 import BeautifulSoup
import re
import random
import pandas as pd
import time
from tqdm import tqdm
```

## 카테고리 링크 크롤링

```python
head = { 'User-Agent': 
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}


url = "https://kin.naver.com/qna/expertAnswerList.naver?dirId=7010101&queryTime=2024-02-07%2016%3A52%3A49&page=1"
link_list2 = []
for page in tqdm(range(1,100)):

        r = requests.get(url.format(page), headers=head)
        bs = BeautifulSoup(r.text)
        data = bs.find("div", class_="qna_list_expertAnswer").find_all("a")

        pattern = r'docId=(\d+)'
        for tag in data:
                tag_str = str(tag)  # 태그 객체를 문자열로 변환
                match = re.search(pattern, tag_str)
                if match:
                        link = match.group(1)
                        link_list2.append(link)
        sleep_time = random.uniform(0,1)
        time.sleep(sleep_time)
```
